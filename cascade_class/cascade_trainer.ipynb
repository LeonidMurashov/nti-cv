```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('test_imgs/img1.jpg')
plt.imshow(img)
```

```python
cascade = cv2.CascadeClassifier('train_v1/cascade.xml')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray)
```

```python
circles = cascade.detectMultiScale(gray)
```

```python
for(x , y, w, h) in circles:
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)
plt.imshow(img)
```
