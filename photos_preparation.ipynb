```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import os

data_path = 'data/templates/'
```

```python
for img_path in os.listdir(data_path):
    name = os.path.join(data_path, img_path)
    img = cv2.imread(name)
    img = 255 - img
    cv2.imwrite(name, img)
```

```python
plt.figure(figsize=(16, 10))
img = cv2.imread(name)
img[-130:, :130] = 0
plt.imshow(img[600:856, 400:656])
```
